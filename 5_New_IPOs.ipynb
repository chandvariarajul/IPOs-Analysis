```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
!pip install yfinance
import yfinance as yf
%matplotlib inline
```

```python
pre = yf.Ticker("PREMIERENE.NS")
pre_share = pre.history(period="max")
pre_share.head(5)
```

```python
uni = yf.Ticker("UNIECOM.NS")
uni_share = uni.history(period="max")
uni_share.head(5)
```

```python
fir = yf.Ticker("FIRSTCRY.NS")
fir_share = fir.history(period="max")
fir_share.head(5)
```

```python
 ola = yf.Ticker("OLAELEC.NS")
 ola_share = ola.history(period="max")
 ola_share.head(5)
```

```python
gal = yf.Ticker("GALAPREC.NS")
gal_share= gal.history(period="max")
gal_share.head(5)
```

```python
pre_share['Open'].plot(label = "Premier Energies ", figsize = (15,7), color="Black")
uni_share['Open'].plot(label = "Unicommerce eSolutions", color="Blue")
fir_share['Open'].plot(label = "Brainbees Solutions(FirstCry)", color= "Red")
ola_share['Open'].plot(label = "Ola Electric", color = "Green")
gal_share['Open'].plot(label = "Gala Precision Engineering", color = "Orange")
plt.title("Opening of new 5 IPOs of 2024 in Indian Stock Exchange")
plt.xlabel("Date")
plt.ylabel("Opening Price")
plt.legend()
plt.show()
```

```python
pre_share['Volume'].plot(label = "Premier Energies ", figsize = (15,7), color="Black")
uni_share['Volume'].plot(label = "Unicommerce eSolutions", color="Blue")
fir_share['Volume'].plot(label = "Brainbees Solutions(FirstCry)", color= "Red")
ola_share['Volume'].plot(label = "Ola Electric", color = "Green")
gal_share['Volume'].plot(label = "Gala Precision Engineering", color = "Orange")
plt.title("Number of share traded of 5 IPOs of 2024 in Indian Stock Exchange")
plt.legend()
plt.show()
```

```python
pre_share['MarketCap'] = pre_share['Volume'] * pre_share['Open']
uni_share['MarketCap'] = uni_share['Volume'] * uni_share['Open']
fir_share['MarketCap'] = fir_share['Volume'] * fir_share['Open']
ola_share['MarketCap'] = ola_share['Volume'] * ola_share['Open']
gal_share['MarketCap'] = gal_share['Volume'] * gal_share['Open']
pre_share['MarketCap'].plot(label = "Premier Energies ", figsize = (15,7), color="Black")
uni_share['MarketCap'].plot(label = "Unicommerce eSolutions", color="Blue")
fir_share['MarketCap'].plot(label = "Brainbees Solutions(FirstCry)", color= "Red")
ola_share['MarketCap'].plot(label = "Ola Electric", color = "Green")
gal_share['MarketCap'].plot(label = "Gala Precision Engineering", color = "Orange")
plt.title("MarketCap of 5 new IPOs of 2024 in Indian Stock Exchange")
plt.legend()
```

```python
pre_share['MA30'] = pre_share['Open'].rolling(30).mean()
pre_share['Open'].plot(figsize=(15,7))
pre_share['MA30'].plot()
plt.title("Average opening for 30 days for Premier Energies")
plt.legend()
```

```python
uni_share['MA30'] = uni_share['Open'].rolling(30).mean()
uni_share['Open'].plot(figsize=(15,7))
uni_share['MA30'].plot()
plt.title("Average opening for 30 days for Unicommerce eSolutions")
plt.legend()
```

```python
fir_share['MA30'] = fir_share['Open'].rolling(30).mean()
fir_share['Open'].plot(figsize=(15,7))
fir_share['MA30'].plot()
plt.title("Average opening for 30 days for Brainbees Solutions(FirstCry)")
plt.legend()
```

```python
ola_share['MA30'] = ola_share['Open'].rolling(30).mean()
ola_share['Open'].plot(figsize=(15,7))
ola_share['MA30'].plot()
plt.title("Average opening for 30 days for Ola Electric")
plt.legend()
```

```python
gal_share['MA30'] = gal_share['Open'].rolling(30).mean()
gal_share['Open'].plot(figsize=(15,7))
gal_share['MA30'].plot()
plt.title("Average opening for 30 days for Gala Precision Engineering")
plt.legend()
```

```python
data = pd.concat([pre_share['Open'], uni_share['Open'], fir_share['Open'], ola_share['Open'], gal_share['Open']], axis=1)
data.columns = ['Premier Energies', 'Unicommerce eSolutions', 'Brainbees Solutions(FirstCry)', 'Ola Electric', 'Gala Precision Engineering']
scatter_matrix(data, figsize=(15,15), hist_kwds={'bins':50})
plt.show()
```

```python
pre_share['returns'] = (pre_share['Close']/pre_share['Close'].shift(1))-1
uni_share['returns'] = (uni_share['Close']/uni_share['Close'].shift(1))-1
fir_share['returns'] = (fir_share['Close']/fir_share['Close'].shift(1))-1
ola_share['returns'] = (ola_share['Close']/ola_share['Close'].shift(1))-1
gal_share['returns'] = (gal_share['Close']/gal_share['Close'].shift(1))-1
pre_share['returns'].hist(bins=100,label="Premier Energies",figsize=(15,7), alpha=0.5, color = 'Black')
uni_share['returns'].hist(bins=100,label="Unicommerce eSolutions", alpha=0.5, color = 'Blue')
fir_share['returns'].hist(bins=100,label="Brainbees Solutions(FirstCry)", alpha=0.5, color = 'Red')
ola_share['returns'].hist(bins=100,label="Ola Electric", alpha=0.5, color = 'Green')
gal_share['returns'].hist(bins=100,label="Gala Precision Engineering", alpha=0.5, color = 'Orange')
plt.title('Percentage Increase in Stocks')
plt.legend()
```
